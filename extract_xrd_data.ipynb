```python
# Import necessary libraries
import pandas as pd
from google.colab import files

# Function to read .ras file
def read_ras_file(file_path):
    with open(file_path, 'rb') as file:
        content = file.read()

    # Decode the content using the appropriate encoding
    decoded_content = content.decode('latin1')  # You can try 'latin1' or other encodings if needed

    # Split the content into lines
    lines = decoded_content.split('\n')

    # Extract the data starting from the 345th line
    data_lines = lines[344:]

    # Parse the data into a DataFrame
    data = []
    for line in data_lines:
        if line.strip():  # Skip empty lines
            parts = line.split()
            if len(parts) >= 2:
                data.append([float(parts[0]), float(parts[1])])

    df = pd.DataFrame(data, columns=['2theta', 'intensity'])
    return df

# Upload .ras files
uploaded = files.upload()

# Process each uploaded file
for file_name in uploaded.keys():
    # Read the .ras file
    df = read_ras_file(file_name)

    # Save the extracted data to a new Excel file
    output_file_path = file_name.replace('.ras', '_extracted.xlsx')
    df.to_excel(output_file_path, index=False)

    print(f"Data from {file_name} successfully extracted and saved to {output_file_path}")
```
